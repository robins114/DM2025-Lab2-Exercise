## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section

# =========================================================
# LOGISTIC REGRESSION + PREPROCESSING + GRIDSEARCH
# =========================================================

import os
import json
import re
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.linear_model import SGDClassifier

import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)

# ==========================================
# Load competition data
# ==========================================
base_path = "/kaggle/input/dm-lab-2-private-competition"

with open(os.path.join(base_path, "final_posts.json"), "r") as f:
    raw_json = json.load(f)

posts = pd.json_normalize(raw_json)
posts = posts.rename(columns={
    "root._source.post.post_id": "id",
    "root._source.post.text": "text"
})
posts = posts[["id", "text"]]

emotions = pd.read_csv(os.path.join(base_path, "emotion.csv"))
splits   = pd.read_csv(os.path.join(base_path, "data_identification.csv"))

df = posts.merge(emotions, on="id", how="left").merge(splits, on="id", how="left")

train_df = df[df["split"] == "train"].copy()
test_df  = df[df["split"] == "test"].copy()

X_train = train_df["text"].fillna("")
y_train = train_df["emotion"]
X_test  = test_df["text"].fillna("")

# ==========================================
# Preprocessing functions (emoji-friendly)
# ==========================================
def clean_keep_emojis(text: str) -> str:
    text = str(text).lower()
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"@\w+", " ", text)
    text = re.sub(r"#\w+", " ", text)
    text = re.sub(r"[0-9]+", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def whitespace_tokenizer(text: str):
    return clean_keep_emojis(text).split()

from nltk.stem import PorterStemmer
ps = PorterStemmer()

def stem_tokenizer_with_emojis(text: str):
    tokens = clean_keep_emojis(text).split()
    return [ps.stem(t) if re.fullmatch(r"[a-z]+", t) else t for t in tokens]

# ==========================================
# Train/validation split
# ==========================================
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train
)

# ==========================================
# TF-IDF preprocessing variants
# ==========================================
preprocessings = {
    "baseline_raw": TfidfVectorizer(
        max_features=25000,
        ngram_range=(1, 2)
    ),
    "clean_only_emoji": TfidfVectorizer(
        max_features=25000,
        ngram_range=(1, 2),
        preprocessor=lambda x: x,
        tokenizer=whitespace_tokenizer,
        token_pattern=None
    ),
    "clean_stopwords_emoji": TfidfVectorizer(
        max_features=25000,
        ngram_range=(1, 2),
        preprocessor=lambda x: x,
        tokenizer=whitespace_tokenizer,
        token_pattern=None,
        stop_words="english"
    ),
    "clean_stem_emoji": TfidfVectorizer(
        max_features=25000,
        ngram_range=(1, 1),
        preprocessor=lambda x: x,
        tokenizer=stem_tokenizer_with_emojis,
        token_pattern=None
    )
}
```

```python
### Add the code related to the model implementation steps in cells inside this section

# Models + hyperparameters
models = {
    "logreg": LogisticRegression(
        max_iter=2000,
        solver="saga",
        class_weight="balanced",
        n_jobs=-1
    ),
    "svm_balanced": LinearSVC(class_weight="balanced"),
    "sgd_svm": SGDClassifier(
        loss="hinge",
        penalty="l2",
        class_weight="balanced",
        max_iter=1500
    )
}

param_grids = {
    "logreg": {"clf__C": [0.5, 1.0, 2.0]},
    "svm_balanced": {"clf__C": [0.5, 1.0, 2.0]},
    "sgd_svm": {"clf__alpha": [1e-4, 1e-5, 1e-6]}
}

# Training over all (preprocessing Ã— model)
results = []

for prep_name, vectorizer in preprocessings.items():
    for model_name, model in models.items():

        pipe = Pipeline([
            ("tfidf", vectorizer),
            ("clf", model)
        ])

        grid = GridSearchCV(
            pipe,
            param_grid=param_grids[model_name],
            scoring="f1_macro",
            cv=3,
            n_jobs=-1,
            verbose=1,
            error_score=0.0
        )

        grid.fit(X_tr, y_tr)

        best = grid.best_estimator_
        val_pred = best.predict(X_val)
        f1 = f1_score(y_val, val_pred, average="macro")

        results.append({
            "preprocessing": prep_name,
            "model": model_name,
            "f1": f1,
            "best_params": grid.best_params_
        })

# pick best model
results_sorted = sorted(results, key=lambda x: x["f1"], reverse=True)
best = results_sorted[0]

best_model_name = best["model"]
best_prep_name  = best["preprocessing"]
best_params     = best["best_params"]

# Retrain best model on full training set
final_vec = preprocessings[best_prep_name]

if best_model_name == "logreg":
    final_clf = LogisticRegression(
        max_iter=2000, solver="saga",
        class_weight="balanced",
        n_jobs=-1,
        C=best_params["clf__C"]
    )
elif best_model_name == "svm_balanced":
    final_clf = LinearSVC(
        class_weight="balanced",
        C=best_params["clf__C"]
    )
else:
    final_clf = SGDClassifier(
        loss="hinge", penalty="l2",
        class_weight="balanced",
        max_iter=1500,
        alpha=best_params["clf__alpha"]
    )

final_pipeline = Pipeline([
    ("tfidf", final_vec),
    ("clf", final_clf)
])

final_pipeline.fit(X_train, y_train)
final_pred = final_pipeline.predict(X_test)

final_df = pd.DataFrame({
    "id": test_df["id"],
    "predicted_emotion": final_pred
})

final_name = f"/kaggle/working/submission_best_{best_model_name}_{best_prep_name}.csv"
final_df.to_csv(final_name, index=False)
print("Saved:", final_name)
```
